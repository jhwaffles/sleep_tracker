```python
def count_restless_periods(movement_str):
    movement_list = [int(x) for x in movement_str]  # Convert string to list of integers
    restless_periods = 0
    in_restless_period = False
    
    for movement in movement_list:
        if movement > 1:  # If movement is above 1, it's a restless period
            if not in_restless_period:
                restless_periods += 1  # Start of a new restless period
                in_restless_period = True
        else:
            in_restless_period = False  # Reset when movement goes to 1 or below

    return restless_periods

# Example usage
movement_30_sec = "11211111111111111111211112211111121111111111111111121111121111111111121111111111111111111111111111111111111111111111111112111111111111111111111321111211112212121211111121111211111111111121112111111211111211111111111111111111131121111111111111121111111111111111111111111111111111211111111111111111111111111111112312111211111112111121111132111111121211212122122112122122122112111121212312111211211211212111111111111111111112111111111111121111111111111111113112111111111111111123211111311111111111131112111111111111111111111122111123212111111112222111211211121111111111112123211111121111111111111131212232121311111111111112222121121112111121211112211111111111111121221211111221112211131111211112111131112112211111111121112122112132111111211111111112311111121111111111111111111111111111112111111121111111211111112111112111111112111122212212211222121112211111111211111111113111321111112111111111123111"
restless_count = count_restless_periods(input)
print(f"Restless Periods: {restless_count}")
```
